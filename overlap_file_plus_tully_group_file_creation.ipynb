```python
overlap_file = read_csv('overlap_file.csv')
t5_file = read_csv('aj511667t5_augmented.csv')

t5_file.rename(columns={'RA2000': 'RA', 'DEC2000': 'Dec'}, inplace=True)

print(overlap_file.head())
print('\n\n\n')
print(t5_file.head())
```

```python
o_select = ['RA', 'Dec', 'Redshift', '2MRS Local Density', 'k_tc', 'h_tc', 'j_tc', 'e_kt', 'e_ht', 'e_jt', 'e_bv', 'r_iso', 'r_ext', 'SDSS Local Density']
overlap_select = overlap_file[o_select]
print(overlap_select.head())
```

```python
#Logrhok -> Logarithm of luminosity density
#LogLK -> Logarithm of intrinsic Ks luminosity
#Ng -> Number of group members from 2MASS catalog
#Vrms -> Line of sight velocity dispersions
t_select = ['RA', 'Dec', 'Logrhok', 'LogLK', 'Ng', 'Vrms']
t5_select = t5_file[t_select]
print(t5_select.head())
```

```python
#Not the best way to solve merging issue don't run cell!!!
overlap_select['RA'] = overlap_select['RA'].round(1)
overlap_select['Dec'] = overlap_select['Dec'].round(1)
print(overlap_select.head())

t5_select['RA'] = t5_select['RA'].round(1)
t5_select['Dec'] = t5_select['Dec'].round(1)
print(t5_select.head())
```

```python
overlap_with_tully_table = pd.merge(overlap_select, t5_select, on = ['RA', 'Dec'])
print(overlap_with_tully_table.head(10))
print('\n', overlap_with_tully_table.shape)
```

```python
import pandas as pd
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
# Ensure RA and Dec columns are numeric
overlap_select['RA'] = pd.to_numeric(overlap_select['RA'], errors='coerce')
overlap_select['Dec'] = pd.to_numeric(overlap_select['Dec'], errors='coerce')
t5_select['RA'] = pd.to_numeric(t5_select['RA'], errors='coerce')
t5_select['Dec'] = pd.to_numeric(t5_select['Dec'], errors='coerce')

# Drop any rows with invalid RA or Dec values
overlap_select = overlap_select.dropna(subset=['RA', 'Dec'])
t5_select = t5_select.dropna(subset=['RA', 'Dec'])

# Convert RA/Dec into SkyCoord objects with explicit units
overlap_coords = SkyCoord(ra=overlap_select['RA'].values * u.deg,
                          dec=overlap_select['Dec'].values * u.deg)

t5_coords = SkyCoord(ra=t5_select['RA'].values * u.deg,
                     dec=t5_select['Dec'].values * u.deg)

# Match coordinates within 1 arcsecond
idx, d2d, _ = t5_coords.match_to_catalog_sky(overlap_coords)
matched = d2d < 1 * u.arcsec  # Tolerance

# Merge matched rows
merged_table = pd.concat([overlap_select.iloc[idx[matched]].reset_index(drop=True),
                          t5_select[matched].reset_index(drop=True)], axis=1)

print(merged_table)
```

```python
print(merged_table.head())
print('\n\n')
print(merged_table.shape)
```

```python
# Check the columns
print(merged_table.columns)

# Drop the second set of 'RA' and 'Dec' columns
merged_table = merged_table.loc[:, ~merged_table.columns.duplicated()]

# Check the result to make sure duplicates are removed
print(merged_table)
```
